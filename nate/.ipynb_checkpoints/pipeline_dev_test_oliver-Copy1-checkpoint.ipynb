```python
# some lists for storing the scans to analyze and the corresponding sample names
cass = '2049'
alph = ['A','B','C','D','E','F']
samps = [] # for storing the sample names
for ii in alph:
    for jj in range(1,9):
        samps.append(str(ii + str(jj)))

# the scans to analyze
scans = ['Scan5615*','Scan5617*','Scan5619*','Scan5621*','Scan5623*','Scan5625*','Scan5627*','Scan5629*',
        'Scan5635*','Scan5637*','Scan5639*','Scan5641*','Scan5643*','Scan5645*','Scan5647*','Scan5649*',
        'Scan5655*','Scan5657*','Scan5659*','Scan5661*','Scan5663*','Scan5665*','Scan5667*','Scan5669*',
         'Scan5675*','Scan5677*','Scan5679*','Scan5681*','Scan5683*','Scan5685*','Scan5687*','Scan5689*',
         'Scan5695*','Scan5697*','Scan5699*','Scan5701*','Scan5703*','Scan5705*','Scan5707*','Scan5709*',
         'Scan5715*','Scan5717*','Scan5719*','Scan5721*','Scan5723*','Scan5725*','Scan5727*','Scan5729*',
        ]
```

```python
# glob the images
# for the test script I'm going to grab only scans 5615 and 5617

res = [] # for storing the q vs I arrays
for ind,samp in enumerate(samps):
    # glob the csv file
    csvs = glob.glob(dataPath + 'Cassette2049_' + samp + '.csv')[0]
    qvi = pd.read_csv(csvs,header=0) # import the data
    res.append([qvi['q'],qvi['I']]) # append q and I to res
    
```

```python
# there's something strange going on with the integration
# the very last value is 0, but everything else has a constant bg intensity
# let's subtract it out

# sort the array by intensity
# find the first non-zero value
# subtract that value from all entires except those equal to zero
subs = []
for ind,entry in enumerate(res):
    bgs = np.sort(entry[1]) # sort the array by intensity value low -> high
    sub = np.array(entry[1] - bgs[1]) # subtract the first nonzero value from the histogram
    sub[-1] = 0 # the last entry will be negative, so set it to zero
    subs.append(sub)
```

```python
# I'm gonna get an estimate for the FWHM of a large peak
# chose what sample to look at
sample = 20
# chose the range of index values that bound the peak
rl = 1029
rh = 1060

# find the max value of the peak
pmax = np.max(subs[sample][rl:rh])
print(pmax)
# find all the values above half of pmax
ptop = np.where(subs[sample][rl:rh] > pmax/2)[0]

# grab the left and right indices bounding the top half of the peak
fwhm = res[sample][0][rl+ptop[-1]+1] - res[sample][0][rl+ptop[0]+1]
print(fwhm)
full_q = fwhm/2.355

plt.figure()
plt.plot(res[sample][0][rl:rh],subs[sample][rl:rh])
plt.plot(res[sample][0][rl+ptop[0]+1:rl+ptop[-1]+1],subs[sample][rl+ptop[0]+1:rl+ptop[-1]+1],'r*')
plt.show()
```

```python
# okay now we can do the signal separation
# define a few parameters, like the theta index range
thstep = 2000


# first put all the intensity arrays into their own list
# i am going to mask out the final few data points because they are strange
exclude = 5
ql = 0 # lower bound of histogram range
qh = thstep-exclude # upper bound of histogram range
ints = np.zeros([qh,len(scans)])

for ind,item in enumerate(subs):
    ints[ql:qh,ind] = item[ql:qh] - np.min(item[ql:qh])
    
# okay now we can do the signal separation
ss = sep.separate_signal(np.transpose(ints),cutoff = .1, threshold = 50, smooth_q = full_q)


low = 0
high = 48

# let's plot all the raw histograms
plt.figure()
for ii in range(low,high):
    # samples 1 and 6 are acting different
    if ii == 1 or ii == 6:
        continue
    plt.plot(res[ii][0][ql:qh],res[ii][1][ql:qh],label=str(ii))
plt.title('Cassette 2049 Raw Data')
plt.xlabel('q (' + r'$\AA^{-1}$)')
plt.ylabel('Intensity')
plt.savefig('cass2049_raw_data.png')

# okay let's plot all the backgrounds that we got
plt.figure()
for ii in range(low,high):
    # samples 1 and 6 are acting different
    if ii == 1 or ii == 6:
        continue
    plt.plot(res[ii][0][ql:qh],ss[0][ii],label=str(ii))
plt.title('Cassette 2049 Slow q')
plt.xlabel('q (' + r'$\AA^{-1}$)')
plt.ylabel('Intensity')
plt.savefig('cass2049_slow_q.png')

# okay let's plot all the backgrounds that we got
plt.figure()
for ii in range(low,high):
    # samples 1 and 6 are acting different
    if ii == 1 or ii == 6:
        continue
    plt.plot(res[ii][0][ql:qh],ss[1][ii],label=str(ii))
plt.title('Cassette 2049 Fast q')
plt.xlabel('q (' + r'$\AA^{-1}$)')
plt.ylabel('Intensity')
plt.savefig('cass2049_fast_q.png')

# okay let's plot all the backgrounds that we got
plt.figure()
for ii in range(low,high):
    # samples 1 and 6 are acting different
    if ii == 1 or ii == 6:
        continue
    plt.plot(res[ii][0][ql:qh],ss[2][ii],label=str(ii))
plt.title('Cassette 2049 Slow T')
plt.xlabel('q (' + r'$\AA^{-1}$)')
plt.ylabel('Intensity')
plt.savefig('cass2049_slow_T.png')

# okay let's plot all the backgrounds that we got
plt.figure()
for ii in range(low,high):
    # samples 1 and 6 are acting different
    if ii == 1 or ii == 6:
        continue
    plt.plot(res[ii][0][ql:qh],ss[3][ii],label=str(ii))
plt.title('Cassette 2049 Fast T')
plt.xlabel('q (' + r'$\AA^{-1}$)')
plt.ylabel('Intensity')
plt.savefig('cass2049_fast_T.png')
```

```python
# let's just make a plot of the minimum, maximum, and median curves
bgmin = np.zeros(qh)
bgmax = np.zeros(qh)
bgmed = np.zeros(qh)

submat = np.zeros([len(scans),qh])

for ii in range(len(scans)):
    # place all of the background subtracted arrays into submat
    submat[ii,:] = ss[0][ii]

# now iterate through submat and find the minimum, maximum, and median values
for ii,item in enumerate(np.transpose(submat[:,])):
    bgmin[ii] = np.min(item)
    bgmax[ii] = np.max(item)
    bgmed[ii] = np.median(item)
    
plt.figure()
plt.plot(res[0][0][ql:qh],bgmin)
plt.plot(res[0][0][ql:qh],bgmax)
plt.plot(res[0][0][ql:qh],bgmed)
plt.show()
```
